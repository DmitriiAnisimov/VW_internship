## open food facts

```python
import requests
import pandas as pd
```

```python
# Set display option to show all columns
pd.set_option('display.max_columns', None)
```

```python
# Пример запроса для поиска продуктов
search_url = 'https://world.openfoodfacts.org/cgi/search.pl'
params = {
    'search_terms': 'chocolate',
    'fields': 'product_name,code,ingredients_text',
    'json': 1
}
response = requests.get(search_url, params=params)
data = response.json()
```

```python
# Извлечение продуктов из JSON
products = data.get('products', [])

# Создание DataFrame
df = pd.DataFrame(products)

# Вывод первых строк DataFrame
df.head()
```

```python
import pandas as pd
import requests

# Пример запроса для конкретного продукта
product_code = '3046920022651'  # Замените на код вашего продукта
product_url = f'https://world.openfoodfacts.org/api/v0/product/{product_code}.json'
response = requests.get(product_url)
data = response.json()

# Извлечение информации о продукте
product_info = data.get('product', {})

# Создание DataFrame с одной строкой
df = pd.DataFrame([product_info])

# Вывод DataFrame
df
```

```python
import pandas as pd
import requests

search_url = 'https://world.openfoodfacts.org/cgi/search.pl'
params = {
    'search_terms': 'chocolate',
    'fields': 'product_name,code,ingredients_text,quantity,packaging',
    'json': 1
}
response = requests.get(search_url, params=params)
data = response.json()

products = data.get('products', [])
df = pd.DataFrame(products)

# Фильтрация и извлечение информации о весе и объеме из полей 'quantity' и 'packaging'
df[['product_name', 'quantity', 'packaging']].head()
```
